```python
customer_value_segment=[]
for i in range(len(df)):
    if df.Customer_Value[i]>25000:
        customer_value_segment.append('High')
    elif df.Customer_Value[i]<=10000:
        customer_value_segment.append('Low')
    else:
        customer_value_segment.append('Medium')
df['Customer_Value_Segment']=customer_value_segment
df.columns
```

```python
df['Avg_Revenue_per_trip']=df['Customer_Value']/df['Buy_Times']
df['Balance_points']=df['Points_Earned']-df['Points_Redeemed']
df.columns
```

```python
from datetime import date

today = date.today()
print(f"Today's date:::{today}")
cust_hist={}
for i in range(len(df)):
    dt=str(df.Recent_Date[i])
    x=date(int(dt[:4]), int(dt[4:6]), int(dt[6:]))-today
    c=df.Customer_ID[i]
    cust_hist[c]=x.days*-1
cust_hist
```

```python
region=df.Last_region.unique()
state=df.Last_state.unique()
city=df.Last_city.unique()
```

```python
total=df.Customer_Value.sum()
print(f"Total worth:::{total}")
region_perc={}
state_perc={}
city_perc={}
for r in region:
    temp=df[df.Last_region==r].Customer_Value.sum()
    region_perc[r]=temp*100/total

for s in state:
    temp=df[df.Last_state==s].Customer_Value.sum()
    state_perc[s]=temp*100/total
    
for c in city:
    temp=df[df.Last_city==c].Customer_Value.sum()
    city_perc[c]=temp*100/total
```

```python
state_count={}
city_count={}
state_avg_purchase={}
city_avg_purchase={}

total_count=df.Customer_ID.count()
total_purchase=df.Buy_Times.sum()
```

```python
for s in state:
    temp=df[df.Last_state==s].Customer_ID.count()
    state_count[s]=temp
    temp=df[df.Last_state==s].Buy_Times.sum()
    state_avg_purchase[s]=temp*100/total_purchase
    
for c in city:
    temp=df[df.Last_city==c].Customer_ID.count()
    city_count[c]=temp
    temp=df[df.Last_city==c].Buy_Times.sum()
    city_avg_purchase[c]=temp*100/total_purchase
```
