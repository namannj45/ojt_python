```python
df = pd.read_csv('/content/drive/MyDrive/Cust_demo.csv')
print(df.shape)
df.head()
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Cust_data.csv')
print(df1.shape)
df1.head()
```

```python
c1=df [(df['Martial_Status'] =='Married') & (df['Own_House']==1) & (df['age'] > 28)]
c1=c1 [['ID','age','Gender','Location']]
print(c1.shape)
c1.head()
```

```python
c1.sort_values(by=['Location', 'age'], ascending=[True, False])
```

```python
df.rename(columns={'NumberOfDependents':'No_of_dependents'}, inplace=True)
df.head()
```

```python
df1.drop_duplicates(keep=False, inplace=True)
df.drop_duplicates(keep=False, inplace=True)
```

```python
df1['No_of_30_plus_DPD']=df1['No_of_30_59_DPD']+df1['No_of_60_89_DPD']+df1['No_of_90_DPD']
df1['No_of_60_plus_DPD']=df1['No_of_60_89_DPD']+df1['No_of_90_DPD']
df1.head()
```

```python
cust_leftjoin=pd.merge(df1,df,on='ID', how='left')
print(cust_leftjoin.shape)
cust_leftjoin.head()
```

```python
cust_rightjoin=pd.merge(df1,df,on='ID', how='right')
print(cust_rightjoin.shape)
cust_rightjoin.head()
```

```python
cust_innerjoin=pd.merge(df1,df,on='ID')
print(cust_innerjoin.shape)
cust_innerjoin.head()
```

```python
cust_outerjoin=pd.merge(df1,df,on='ID', how='outer')
print(cust_outerjoin.shape)
cust_outerjoin.head()
```

```python
c11=df.sample(frac=0.05)
c22=df.sample(n=10000)
print(f"sample 1 shape:::{c11.shape}")
print(f"sample 2 shape:::{c22.shape}")
```

```python
no_of_customers=len(cust_leftjoin)
no_of_married_customers=len(cust_leftjoin[cust_leftjoin['Martial_Status']=='Married'])
print(f"No. of Customers={no_of_customers}")
print(f"No. of Married Customers={no_of_married_customers}")
```

```python
m=len(cust_innerjoin[cust_innerjoin['Gender']==0])
f=len(cust_innerjoin[cust_innerjoin['Gender']==1])
t=len(cust_innerjoin)
print(f"Male Customers are {m}")
print(f"Female Customers are {f}")
print(f"Male Customers percentage is {round(m*100/t,2)}%")
print(f"Female Customers percentage is {round(f*100/t,2)}%")
```

```python
revm=cust_innerjoin[cust_innerjoin['Gender']==0]['RevolvingUtilization'].mean()
revf=cust_innerjoin[cust_innerjoin['Gender']==1]['RevolvingUtilization'].mean()
revm_std=cust_innerjoin[cust_innerjoin['Gender']==0]['RevolvingUtilization'].std()
revf_std=cust_innerjoin[cust_innerjoin['Gender']==1]['RevolvingUtilization'].std()
print(f"Male avg Revolving Utilization:::{round(revm,2)}")
print(f"Female avg Revolving Utilization:::{round(revf,2)}")
print(f"Male Standard Deviation of Revolving Utilization:::{round(revm_std,2)}")
print(f"Female Standard Deviation of Revolving Utilization:::{round(revf_std,2)}")
```

```python
inc_m=cust_innerjoin[cust_innerjoin['Gender']==0]['MonthlyIncome'].mean()
inc_f=cust_innerjoin[cust_innerjoin['Gender']==1]['MonthlyIncome'].mean()
max_inc_m=cust_innerjoin[cust_innerjoin['Gender']==0]['MonthlyIncome'].max()
max_inc_f=cust_innerjoin[cust_innerjoin['Gender']==1]['MonthlyIncome'].max()
std_inc_m=cust_innerjoin[cust_innerjoin['Gender']==0]['MonthlyIncome'].std()
std_inc_f=cust_innerjoin[cust_innerjoin['Gender']==1]['MonthlyIncome'].std()

print(f"Average Male Monthly Income is {round(inc_m,2)}")
print(f"Average Female Monthly Income is {round(inc_f,2)}")
print(f"Max Male Monthly Income is {round(max_inc_m,2)}")
print(f"Max Female Monthly Income is {round(max_inc_f,2)}")
print(f"Standard Deviation in Male Monthly Income is {round(std_inc_m,2)}")
print(f"Standard Deviation in Female Monthly Income is {round(std_inc_f,2)}")
```

```python
age_m=cust_innerjoin[cust_innerjoin['Gender']==0]['age'].mean()
age_f=cust_innerjoin[cust_innerjoin['Gender']==1]['age'].mean()
dep_m=cust_innerjoin[cust_innerjoin['Gender']==0]['No_of_dependents'].mean()
dep_f=cust_innerjoin[cust_innerjoin['Gender']==1]['No_of_dependents'].mean()

print(f"Male Average Age is {round(age_m,2)}")
print(f"Female Average Age is {round(age_f,2)}")
print(f"Average No. of dependents on male customer is {round(dep_m,2)}")
print(f"Average No. of dependents on female customer is {round(dep_f,2)}")
```

```python
m_dlq=len(cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0])
f_dlq=len(cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1])
t=len(cust_innerjoin)
print(f"delinquent Customers are {m_dlq}")
print(f"non-delinquent Customers are {f_dlq}")
print(f"delinquent Customers percentage is {round(m_dlq*100/t,2)}%")
print(f"non-delinquent Customers percentage is {round(f_dlq*100/t,2)}%")

rev_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['RevolvingUtilization'].mean()
rev_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['RevolvingUtilization'].mean()
rev_dlq_std=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['RevolvingUtilization'].std()
rev_non_dlq_std=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['RevolvingUtilization'].std()
print(f"delinquent avg Revolving Utilization:::{round(rev_dlq,2)}")
print(f"non-delinquent avg Revolving Utilization:::{round(rev_non_dlq,2)}")
print(f"delinquent Standard Deviation of Revolving Utilization:::{round(rev_dlq_std,2)}")
print(f"non-delinquent Standard Deviation of Revolving Utilization:::{round(rev_non_dlq_std,2)}")

inc_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['MonthlyIncome'].mean()
inc_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['MonthlyIncome'].mean()
max_inc_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['MonthlyIncome'].max()
max_inc_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['MonthlyIncome'].max()
std_inc_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['MonthlyIncome'].std()
std_inc_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['MonthlyIncome'].std()

print(f"Average delinquent Monthly Income is {round(inc_dlq,2)}")
print(f"Average non-delinquent Monthly Income is {round(inc_non_dlq,2)}")
print(f"Max delinquent Monthly Income is {round(max_inc_dlq,2)}")
print(f"Max non-delinquent Monthly Income is {round(max_inc_non_dlq,2)}")
print(f"Standard Deviation in delinquent Monthly Income is {round(std_inc_dlq,2)}")
print(f"Standard Deviation in non-delinquent Monthly Income is {round(std_inc_non_dlq,2)}")

age_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['age'].mean()
age_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['age'].mean()
dep_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==0]['No_of_dependents'].mean()
dep_non_dlq=cust_innerjoin[cust_innerjoin['SeriousDlqin2yrs']==1]['No_of_dependents'].mean()

print(f"delinquent Average Age is {round(age_dlq,2)}")
print(f"non-delinquent Average Age is {round(age_non_dlq,2)}")
print(f"Average No. of dependents on delinquent customer is {round(dep_dlq,2)}")
print(f"Average No. of dependents on non-delinquent customer is {round(dep_non_dlq,2)}")
```
